```python
import pandas as pd
import numpy as np
import joblib
```

```python
df = pd.read_csv("customer.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
import math
median_SpendingScore = math.floor(df.SpendingScore.median())
median_SpendingScore
```

```python
df.SpendingScore = df.SpendingScore.fillna(median_SpendingScore)
df
```

```python
dummies = pd.get_dummies(df['Gender'], dtype=int)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
df = merged.drop(['Gender','Female'],axis='columns')
df
```

```python
df.isnull().sum()
```

```python
features = ['Age','AnnualIncome','NumberOfPurchases',
            'TimeSpentOnWebsite','LastPurchaseDaysAgo','Male']

x = df[features]
y = df['SpendingScore']

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

model = LinearRegression()

model.fit(x_train, y_train)

#joblib.dump(sgd, r"C:/Users/srivi/Downloads/linear_model.pkl")
# print("Model saved successfully")

loaded_model = joblib.load(
    r"C:/Users/srivi/Downloads/linear_model.pkl"
)

y_pred = loaded_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MSE :", mse)
print("RMSE:", rmse)
print("R2  :", r2)
```

```python
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create scaler
scaler = StandardScaler()

# Fit scaler ONLY on training data
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Train model
model = LinearRegression()
model.fit(x_train_scaled, y_train)

# Save both model and scaler
joblib.dump(model, "linear_model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("Model and scaler saved successfully!")
```

```python
import joblib
import numpy as np

# Load saved files
model = joblib.load("linear_model.pkl")
scaler = joblib.load("scaler.pkl")

# New customer data (same order as training)
# ['Age','AnnualIncome','NumberOfPurchases',
#  'TimeSpentOnWebsite','LastPurchaseDaysAgo','Male']

new_data = np.array([[30, 60000, 15, 35, 5, 1]])

# Scale using saved scaler (IMPORTANT)
new_data_scaled = scaler.transform(new_data)

# Predict
prediction = model.predict(new_data_scaled)

print("Predicted Spending Score:", prediction[0])
```
